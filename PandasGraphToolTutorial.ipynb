```python
node_id_mapping = defaultdict(lambda :graph.add_vertex())
```

### now we need a function which adds edges

```python
def process_row(data,node_id_mapping=None):
    source, destination = data
    #resolve id to vertex
    source_vertex = node_id_mapping[source]
    destination_vertex = node_id_mapping[destination]
    #edge already exists?
    edge = graph.edge(source_vertex,destination_vertex)
    if edge is not None:
        #yes: return it
        return edge
    else:
        #no: create one and create it
        return graph.add_edge(source_vertex,destination_vertex)
```

### we need the answers dataset and filter out self-answers

```python
answers = answers.query('source != destination')
print answers.head()
```

### now get those edges!

```python
answers['edge'] = answers[['source','destination']].apply(process_row,axis=1,node_id_mapping = node_id_mapping)
```
